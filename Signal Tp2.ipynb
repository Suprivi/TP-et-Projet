```python
def sinusoide(Fs, Fe, duree):
    N = Fe*duree
    tableau = np.zeros(N)
    for i in range(N):
        tableau[i] = np.sin(2*np.pi*i*Fs/Fe)
    return tableau

S = sinusoide(440,30000,3)
save_wave(S,'Question1.wav',30000)
play_wave('Question1.wav')
```

```python
#3)a
print(fft.fft(S))

plt.plot(x,np.abs(fft.fft(S)))
plt.show()
```
